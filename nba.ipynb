Kot smo že prej omenili, si bomo sedaj pogledali in primerjali pobližje dve ekipi: Golden State Warriors in Miami Heat.
Začeli bomo z Golden State.

#### Golden State Warriors

```python
Golden_state = nba[nba.ekipa == 'GS'].sort_values('toèke_na_tekmo', ascending=False)
Golden_state
```

Kot je razvidno iz spodnje tabele, sta Stephen Curry in Klay Thompson najboljša igralca v tej ekipi-oba namreč prevladujeta v kategoriji 'odigrane igre', 'minute na igro' ter 'točke na tekmo'.

```python
najboljši_igralci_G = Golden_state[(Golden_state.toèke_na_tekmo >20) & (Golden_state.odigrane_igre>75) & (Golden_state.prosti_meti > 800)]
najboljši_igralci_G
```

Sedaj dodajmo v našo Golden State tabelo dodatna stolpca z imenom 'plača', kjer je podatek o plači v tej sezoni, ter 'starost', kjer bo podatek o letih igralcev.

```python
Golden_state['plača'] = ['$34,382,550','$25,000,000','$17,826,150', '$16,400,000', '$14,814,815', '-', '$1,471,382', '$3,477,600', '$7,692,308', '$1,471,382', '$1,312,611']
Golden_state['starost'] = [29, 28, 27, 27, 33, 0, 29, 33, 31, 36, 21]
Golden_state

```

Kot vidimo najboljši seveda zaslužijo največ. Med vodilnimi sta Stephen Curry, Klay Thompson ter Kevin Durant, kateri tudi je med najboljšimi igralci GS, ampak ga v tabeli_ ne vidimo, saj se je to sezono poškodoval.
Plače se zelo razlikujejo, tu znaša razlika med najvišjo in najnižjo kar $33,069,939.

Izračunajmo še povprečno starost:

```python
Vsota = Golden_state['starost'].sum()
Vsota / 10
```

### Miami Heat

```python
Miami = nba[nba.ekipa == 'MIA'].sort_values('toèke_na_tekmo', ascending=False)
Miami
```

Kot zgoraj tudi tu poglejmo, kdo je najboljši igralec. A za razliko od prej, moramo tu znižati kriterije, saj če pustimo enake kot zgoraj, potem noben igralec ne ustreza. Torej že takoj vidimo razlike v uspešnosti ekipe.

```python
najboljši_igralci_M = Miami[(Miami.toèke_na_tekmo >16) & (Miami.odigrane_igre>66) & (Miami.prosti_meti > 500)]
najboljši_igralci_M
```

Vodilna sta Hassan Whiteside in seveda poznani Goran Dragic.

Zopet bomo dodali dva dodatna stolpca, analogno kot zgoraj.

```python
Miami['plača'] = ['$17,000,450', '$23,775,506', '$5,881,260', '$13,734,000', '$6,270,000', '$1,312,611', '-', '-']
Miami['starost'] = [31, 28, 25, 30, 29, 26, 0,0]
Miami
```

Tudi tu sta pri plačah dominantna najboljša igralca. Presenetljivo, da ima Hassan Whiteside višjo plačo,a vendar je v vseh kategorijah slabši od Gorana Dragića.


Izračunajmo še povprečno starost:

```python
VsotaM = Miami['starost'].sum()
VsotaM / 6
```

Ni pravilo, ampak mlajši igralci imajo načeloma nižje plače kot starejši. Slednji imajo višje zaradi boljšega statusa, daljše kariere in s tem večjega izkoristka svojih sposobnosti.

Za zanimovost si poglejmo najboljše igralce obeh ekip skupaj.
Seveda sta oba najboljša igralca Golden State-a boljša od drugih dveh, a se med seboj ne razlikuejo veliko. Pri odigranih igrah, minutah na tekmo, metu iz igre ter pri trojkah so razlike minimalne, medtem ko pa se malo več razlikujejo pri prostih metih ter pri pomembni kategoriji točke na tekmo, katero tudi prikazuje spodnji graf.

```python
tabeli = [najboljši_igralci_G, najboljši_igralci_M]

skupna_tabela = pd.concat(tabeli)
skupna_tabela
```

```python
skupna_tabela['toèke_na_tekmo'].plot(kind='pie')
igralec = ['Stephen Curry', 'Klay Thompson', 'Goran Dragić', 'Hassan Whiteside']
plt.legend(labels = igralec)
plt.show()
```

Vrnimo se sedaj nazaj k skupni analizi vseh igralcev.

Primerjajmo procente vseh različnih metov hkrati. Ampak to nam ne da realne slike o najboljšem strelcu, saj je povprečje metov posledica odigranih iger in minut na tekmo.
